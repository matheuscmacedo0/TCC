## Fase de transformação dos dados

Nesta fase é feita a transformação dos dados, de forma que fiquem em conformidade com os padrões exijidos pelo algoritmo de mineração de dados.
Neste estudo de caso, será utilizado o algoritmo Apriori para mineração de regras de associação.
Para este tipo de técnica, é preferível que a base de dados contenha dados qualitativos, ou seja, que o dado seja uma característica, e não quantitativas, apenas numéricos.

```python
import pandas as pd
```

```python
df = pd.read_csv('alunos_categoria_administrativa.csv', sep='|')
```

#### Transformando os atributos nominalmente

#### Turno

```python
df['TP_TURNO'] = df['TP_TURNO'].apply(lambda x: 'matutino' if x==1.0 else('vespertino' if x == 2.0 else('noturno' if x == 3.0 else('integral' if x == 4.0 else('EaD')))))
```

#### Data de ingresso para Ano Ingresso

```python
df['DT_INGRESSO_CURSO'] =  pd.to_datetime(df['DT_INGRESSO_CURSO'], format='%d/%m/%Y')
df['ANO_INGRESSO'] = df['DT_INGRESSO_CURSO'].dt.year
```

#### Criando IDADE_INGRESSO

```python
df['IDADE_INGRESSO'] = df['ANO_INGRESSO'].sub(df['NU_ANO_NASCIMENTO'])
```

```python
df
```

#### IDADE_INGRESSO

```python
df['IDADE_INGRESSO_N'] = df['IDADE_INGRESSO'].apply(lambda x: 'MENOR_19' if x<19 else('MENOR_23' if x <23 else('MENOR_27' if x < 27 else('MAIOR_27') )))
```

#### SEXO

```python
df['SEXO_N'] = df['TP_SEXO'].apply(lambda x: 'FEMININO' if x == 1 else('MASCULINO'))
```

#### ETNIA

```python
df['ETNIA_N'] = df['TP_COR_RACA'].apply(lambda x: 'BRANCA' if x == 1 else('PRETA' if x == 2 else('PARDA' if x == 3 else('AMARELA' if x == 4 else('INDIGENA' if x == 5 else('NAO_DECLARADO'))))))
```

#### IN_RESERVA_ETNICO

```python
df['IN_RESERVA_ETNICO_N'] = df['IN_RESERVA_ETNICO'].apply(lambda x: 'SIM' if x == 1 else('NAO'))
```

#### IN_RESERVA_RENDA_FAMILIAR

```python
df['IN_RESERVA_RENDA_FAMILIAR_N'] = df['IN_RESERVA_RENDA_FAMILIAR'].apply(lambda x: 'SIM' if x == 1 else('NAO'))
```

#### CONCLUINTE

```python
df['CONCLUINTE_N'] = df['IN_CONCLUINTE'].apply(lambda x: 'SIM' if x == 1 else('NAO'))
```

#### Eliminando os atributos numéricos que foram usados para transformação

```python
df.drop(['TP_GRAU_ACADEMICO', 'TP_CATEGORIA_ADMINISTRATIVA', 'TP_ORGANIZACAO_ACADEMICA', 'DT_INGRESSO_CURSO', 'IDADE_INGRESSO', 'TP_SEXO', 'TP_COR_RACA', 'IN_RESERVA_ETNICO', 'IN_RESERVA_RENDA_FAMILIAR', 'IN_CONCLUINTE', 'NU_MES_NASCIMENTO', 'NU_DIA_NASCIMENTO', 'NU_ANO_NASCIMENTO', 'ANO_INGRESSO'], axis=1, inplace=True)
```

#### Exportando a base transformada

```python
df.to_csv('alunos_done.CSV', sep=',', index=None)
```
